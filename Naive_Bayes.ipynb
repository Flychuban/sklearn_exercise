```python
from sklearn.datasets import load_wine
import pandas as pd

dataset = load_wine()

df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)

df["result"] = dataset.target
df.head(100)
```

```python
y = df["result"]
X = df.drop(["result"], axis=1, inplace=False)
X.head()
```

Scale all values

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X[X.columns] = scaler.fit_transform(X)
X.head()
```

```python
y.value_counts()
```

```python
# Train - Test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB

model_MulNB = MultinomialNB()
model_GauNB = GaussianNB()

model_MulNB.fit(X_train, y_train)
model_GauNB.fit(X_train, y_train)

print(f"Multinomial score: {model_MulNB.score(X_test, y_test)}")
print(f"Gaussian score:{model_GauNB.score(X_test, y_test)}")
```

```python
predictions_mul = model_MulNB.predict(X_test[:10])
predictions_gau = model_GauNB.predict(X_test[:10])
predictions_actual = []

for val in y_test[:10]:
  predictions_actual.append(val)

print(f"Actual          : {predictions_actual}")
print(f"Multinomial pred: {predictions_mul}")
print(f"Gaussian pred   : {predictions_gau}")
```

Performance is good. NultinomialNB score is same as Gaussian Score
